### Erster DF: Strains -> Basis Informationen


```python
# Assign & Rename

strains = df[['id','name','subtitle','cannabinoid_thc_percentile50','category','averageRating' ,'topEffect','parent_slugs', 'children_slugs']].rename(
    columns={'name': 'strain','cannabinoid_thc_percentile50': 'thc_median','parent_slugs':'parents', 'children_slugs':'children'}
)
strains = strains.set_index('id')
strains.head(3)
```

```python
strains.sort_values('thc_median', ascending=False).head()
```

```python
strains[strains['strain'].str.startswith('Animal')].sort_values('strain')
```

## Überblick


```python
# THC Verteilung

fig = px.histogram(strains, x='thc_median', nbins=80, title="Verteilung des THC-Median")
fig.show()
```

```python
# Durschn. Rating nach Kategorie

fig = px.box(strains, x='category', y='averageRating', title="Rating nach Kategorie")
fig.show()
```

```python
# Durschn. Rating nach bestem Effekt

fig = px.box(strains, x='topEffect', y='averageRating', title="Rating nach Top-Effect")
fig.show()
```

```python
avg = strains[strains['averageRating'] == 1]
print(avg[avg['topEffect'] =='Sleepy']['strain'])
print(avg[avg['topEffect'] =='Focused']['strain'])
print(avg[avg['topEffect'] =='Talkative']['strain'])
print(avg[avg['topEffect'] =='Aroused']['strain'])
```

```python
# Auftreten der Top-Effekte
fig = px.histogram(strains, x='topEffect',  title="Verteilung des Top-Effekts")
fig.show()
```

```python
fig = px.scatter(strains, x="thc_median", y="averageRating",
                 color="category",
                 hover_data=["strain"],
                 title="THC-Median vs. Rating (farblich nach Kategorie)")
fig.show()
```
