```python
import pandas as pd
```

Untersucht wird ein Dataset von Hugginface.co (Apache license 2.0)

```python
strains = pd.read_csv('./csv/strains.csv')
strains.head()
```

```python
strains.describe()
```

```python
ratings = strains[[ 'name','averageRating' ]]
desc_strains = ratings.sort_values('averageRating', ascending=False)
desc_strains.to_csv('desc_strains.txt', index=True)
```

```python
ratings = strains[[ 'name','averageRating' ]]
ratings[ratings["averageRating"]==5].to_csv('top_strains.txt', index=True)
```
